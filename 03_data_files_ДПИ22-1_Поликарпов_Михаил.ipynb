### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
jobs = data['jobs'].explode('jobs').unique()
job_people = {}
for job in sorted(jobs):
    names = data[data['jobs'].apply(lambda array: job in array)]['username'].tolist()
    job_people[job] = sorted(names)
job_people
````

2.2 Сохраните результаты в файл `job_people.pickle` и в файл `job_people.json` с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент `indent`.

````python
import pickle
import os

with open('data\job_people.pickle', 'wb') as job_data:
    pickle.dump(job_people, job_data)
    
with open('data\job_people.json', 'w') as job_data:
    json.dump(job_people, job_data, indent=2)
    
size_pickle = round(os.path.getsize("data\job_people.pickle")/1024)
size_json = round(os.path.getsize("data\job_people.json")/1024)
    
print(f'Вес pickle-файла: {size_pickle}Kb')
print(f'Вес json-файла: {size_json}Kb')
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
from bs4 import BeautifulSoup

with open('data\steps_sample.xml', 'r') as xml_data:
    steps_sample = xml_data.read()
print('Чтение файла закончено')    
Bs_data = BeautifulSoup(steps_sample, "xml")
````

````python
import json
recipes_steps = {}
for recipe in Bs_data.find_all('recipe'):
    name_id = int(recipe.find('id').text)
    recipes_steps[name_id] = [step.text for step in recipe.find_all('step')]
    
with open('data\steps_sample.json', 'w') as json_data:
    json.dump(recipes_steps, json_data, indent=2)
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
step_id = {}
for recipe in Bs_data.find_all('recipe'):
    len_step_recipe = len(recipe.find_all('step'))
    step_id[len_step_recipe] = step_id.get(len_step_recipe, []) + [int(recipe.find('id').text)]
````
